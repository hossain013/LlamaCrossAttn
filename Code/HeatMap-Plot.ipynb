```
import pandas as pd

# Averaged results (computed across folds)
data = {
    "Model": ["ESM2", "ProtT5", "SeqVec"],
    "accuracy": [0.88247, 0.86171, 0.87084],
    "f1": [0.88305, 0.86335, 0.87290],
    "recall": [0.88749, 0.87362, 0.88756],
    "precision": [0.87920, 0.85412, 0.85964],
    "roc_auc": [0.94242, 0.92882, 0.93478],
    "aupr": [0.94163, 0.91835, 0.92845],
    "specificity": [0.87785, 0.84983, 0.85431],
}

df_avg = pd.DataFrame(data).sort_values(by="accuracy",ascending = False)
df_avg.columns=df_avg.columns.str.title().str.replace("_"," ")
df_avg.set_index("Model", inplace=True)

print(df_avg)
```

```
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

sns.heatmap(
    df_avg,
    annot=True,
    fmt=".3f",
    cmap="coolwarm",
    linewidths=0.5,
    cbar_kws={"label": "Score"}
)

# plt.xlabel("Metrics", fontsize=12)
plt.ylabel("Embedding Model", fontsize=12)
plt.title("Performance Comparison of Embedding Models", fontsize=14)

plt.tight_layout()
plt.show()
```
